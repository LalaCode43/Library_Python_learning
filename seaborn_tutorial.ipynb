```python
# load dataset 
cereal_df = pd.read_csv('./data/cereal.csv')
```

```python
cereal_df.head()
```

```python
# i rrelevant fields

fileds = ['shelf', 'weight', 'cups', 'rating']
cereal_df_new = cereal_df.drop(fileds, axis=1)

cereal_df_new.head()
```

```python
# corr()

ceral_corr = cereal_df_new.corr()
ceral_corr
```

```python
# ones_like 
ones_corr = np.ones_like(ceral_corr, dtype='bool')
ones_corr
```

```python
#
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=ceral_corr, mask=mask)
```

```python
adjusted_cereal_corr = ceral_corr.iloc[1:, :-1]
adjusted_mask = mask[1:, :-1]
```

```python
fix, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(data=adjusted_cereal_corr,
            mask=adjusted_mask,
            annot=True, fmt='.2f',
            cmap='Blues',
            vmin=-1, vmax=1,
            linecolor='white', linewidths=1, annot_kws={'fontsize':13})

yticks = [i.upper() for i in adjusted_cereal_corr.index]
xticks = [i.upper() for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks, rotation=0, fontsize=13);
ax.set_xticklabels(xticks, rotation=30, fontsize=13);
title = 'CORRELATION MATRIX\nSAMPLED CEREALS COMPOSITION\n'
ax.set_title(title, fontsize=18, loc='left');
```
