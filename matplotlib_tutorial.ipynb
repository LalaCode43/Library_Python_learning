## 1. Pyplot API vs OOP API
* Pyplot API -> quick
* OOP API -> advanced

### Pyplot API

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), c='b', linestyle="dashed", label='sin(x)');
plt.plot(x, np.cos(x), c='r', label='cos(x)')
plt.title('Sin(x) and Cos(x)')
plt.xlabel('Bien x');
plt.ylabel('f(x)');

## set limit of the axis
# 1:
# plt.xlim([0, 4])
# plt.ylim([-0.7, 0.6])
# 2:
#plt.axis([0, 0.4, -0.7, 0.6]);

# plt.axis('equal')
# plt.axis('tight')

plt.legend(); # display labels
```

### Oject-oriented API

```python
# 1 prepare data
x = np.array([1, 2, 3, 4])
y = np.array([11, 22, 33, 44])

#2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5))

#3. plot data
ax.plot(x, y);

# ax.set(title='A simple plot',
#        xlabel='x',
#        ylabel='y');  
ax.set_title('A simple plot');
ax.set_xlabel('x');
ax.set_ylabel('y');
```

### 2. Scatter

```python
plt.scatter(x, np.exp(x));
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x, y, 
                  s=sizes,
                  c=colors,
                  cmap='viridis',
                  alpha=0.5);
fig.colorbar(img1);

```

### 3. Bar char
* Vertical
* Horizontal

```python
soft_drinks_prices ={
    'Coke': 10,
    'Pessi': 12,
    'Sprite': 8
}

fig, ax = plt.subplots()
ax.bar(soft_drinks_prices.keys(), soft_drinks_prices.values());
ax.set(
    title="Bach hoa xanh's drink prices",
    ylabel='Price ($)'
);
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drinks_prices.keys()), list(soft_drinks_prices.values()));
ax.set(
    title="Bach hoa xanh's drink prices",
    xlabel='Price ($)'
);
```

### 4. Histogram

```python
np.random.seed(4)
student_height = np.random.normal(170, 10, 250)
student_height[:10], len(student_height)
```

```python
plt.hist(student_height, bins=30);
```

### 5. Subplots

```python
# Option 1:
fig, ((ax1, ax2),  (ax3, ax4)) = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10, 8)
)

# plot axis 1
ax1.plot(x, x/2);
ax1.plot(x, x*2)

# plot axis 2
ax2.scatter(
    np.random.random(50), np.random.random(50), 
    alpha=0.6, c=rng.rand(50), cmap='viridis'
);

# plot axis 3
ax3.hist(student_height, bins=30);

# plot axis 4
ax4.bar(soft_drinks_prices.keys(), soft_drinks_prices.values());
ax4.set(
    title="Bach hoa xanh's drink prices",
    ylabel='Price ($)'
);
```

```python
# Option 2:
fig, ax = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10, 8)
)

# plot axis 1
ax[0, 0].plot(x, x/2);
ax[0, 0].plot(x, x*2)

# plot axis 2
ax[0, 1].scatter(
    np.random.random(50), np.random.random(50), 
    alpha=0.6, c=rng.rand(50), cmap='viridis'
);

# plot axis 3
ax[1, 0].hist(student_height, bins=30);

# plot axis 4
ax[1, 1].bar(soft_drinks_prices.keys(), soft_drinks_prices.values());
ax[1, 1].set(
    title="drink's prices",
    ylabel='Price ($)'
);
```
